```python

import pandas as pd

cities_df = pd.read_csv(
    "../data/communicipalities.csv",
    sep=";"
)
# print(cities_df.columns.tolist())
# cities_df.head()
```

```python

count = cities_df["NIS-code Municipaity"].nunique()
print(cities_df["NIS-code Municipaity"].unique())
print(f"Number of unique municipalities: {count}")
```

```python
NIS_PROVINCE_MAP = {
    "11": "Antwerp",
    "12": "Antwerp",

    "21": "Brussels-Capital",

    "23": "Flemish Brabant",
    "24": "Flemish Brabant",

    "25": "Walloon Brabant",

    "31": "West Flanders",
    "32": "West Flanders",
    "33": "West Flanders",
    "34": "West Flanders",

    "35": "East Flanders",
    "36": "East Flanders",
    "37": "East Flanders",
    "38": "East Flanders",
    "39": "East Flanders",

    "41": "Hainaut",
    "42": "Hainaut",
    "43": "Hainaut",
    "44": "Hainaut",
    "45": "Hainaut",

    "51": "Liège",
    "52": "Liège",
    "53": "Liège",

    "56": "Luxembourg",
    "57": "Luxembourg",

    "61": "Namur",
    "62": "Namur",

    "71": "Limburg",
    "72": "Limburg",

    "81": "Limburg",
    "82": "Luxembourg",
    "83": "Luxembourg",
    "84": "Luxembourg",
    "85": "Luxembourg",
    "91": "Namur",
    "92": "Namur",
    "93": "Namur",
}

def get_province_from_nis(nis):
    prefix = str(nis)[:2]
    return NIS_PROVINCE_MAP.get(prefix)


cities_df["province"] = cities_df["NIS-code Municipaity"].apply(
    get_province_from_nis
)

result = cities_df[
    [
        "Postal code",
        "NIS-code Municipaity",
        "province",
        "Municipality_NL"
    ]
].drop_duplicates()

print(result)
```
